```python

# Switching from days to seconds on the x axis


seconds_t_array=np.arange(0,10*u.day.to(u.s), 20)

seconds_delta_t=seconds_t_array[1]-seconds_t_array[0]

def trial(fwhm_array):
    
    dist_to_nans=np.zeros(len(fwhm_array))
    
    for i in range(len(fwhm_array)):
        flux=flare_model(seconds_t_array, 5*u.day.to(u.s), fwhm_array[i], .01)
        
        dist_to_nans[i]=(np.where(flux>0)[0][0] - len(flux[np.isnan(flux)]))* seconds_delta_t / fwhm_array[i]
        
    return dist_to_nans



fwhm_array=np.arange(min(lupita_tab['FWHM'])*u.day.to(u.s), max(lupita_tab['FWHM'])*u.day.to(u.s), 10)

fwhm_array

trial(fwhm_array)
```
